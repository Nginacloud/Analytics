```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Distribution of Temperature and Humidity

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Temp_C'], kde=True)
plt.title('Distribution of Temperature')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Rel Hum_%'], kde=True)
plt.title('Distribution of Humidity')
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.show()
```

A time series analysis
How temperaature varies over time - graph suggests a typical seasonal pattern


```python
plt.figure(figsize=(14, 7))
df['Date/Time'] = pd.to_datetime(df['Date/Time'])
df.set_index('Date/Time', inplace=True)
df['Temp_C'].plot()
plt.title('Temperature Over Time')
plt.xlabel('Date/Time')
plt.ylabel('Temperature')
plt.show()
```

Time series for Wind speed over time - theres occassional peaks within the frequent fluctuations inn wind speed.
there are few instances where wind spikes over (0 - 20) which is the majority bracket.

```python
plt.figure(figsize=(14, 7))
df['Wind Speed_km/h'].plot()
plt.title('Wind Speed over Time')
plt.xlabel('Date')
plt.ylabel('Wind Speed km/h')
plt.show()
```

Correlation Matrix / heatmap - the relationship between these aspects of weather
Red - (closer to 1) represent positive correlation 
Blue - (closer to -1) represent negative correlation
white shades - little to no correlation

```python
correlation_matrix = df.corr()
plt.figure(figsize=(14, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

Seasonal variation of the weather category

```python
print(df)
```

occurences of each weather type per month

```python
monthly_weather = df.pivot_table(index=df['Date/Time'].dt.month, columns='Weather', aggfunc='size', fill_value=0)
print(monthly_weather)
```

```python
monthly_weather.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Weather Conditions by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Weather Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
print(df)
```

```python
mean_weather = df.groupby('Weather').mean().reset_index()
print(mean_weather)
```

```python
df['month'] = df['Date/Time'].dt.month
monthly_avg_temp = df.groupby('month')['Temp_C'].mean()
print(monthly_avg_temp)
```

```python
plt.figure(figsize=(10, 6))
monthly_avg_temp.plot(kind='bar')
plt.title('Average Monthly Temperature')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.xticks(rotation=0)
plt.show()
```
